# 시간별 데이터 사용

```python
df_daily=df
df_daily.head(32)
```

```python
df_daily = df_daily.dropna(subset=['Seoul_PM2.5','Seoul_PM10']).reset_index(drop = True) 
#target인 Seoul_PM2.5/10에 nan 값 있는 행 삭제, index 초기화

pm25 = [16, 36, 76]
pm10 = [31, 81, 151]

df_daily['class_pm2.5'] = -1
df_daily['class_pm10'] = -1

for i,pm in enumerate(pm25):
  print(i,pm)
  df_daily.loc[(df_daily['class_pm2.5'] == -1) & (df_daily['Seoul_PM2.5'] < pm), 'class_pm2.5'] = i
df_daily.loc[(df_daily['class_pm2.5'] == -1), 'class_pm2.5'] = 3

for i,pm in enumerate(pm10):
  print(i,pm)
  df_daily.loc[(df_daily['class_pm10'] == -1) & (df_daily['Seoul_PM10'] < pm), 'class_pm10'] = i
df_daily.loc[(df_daily['class_pm10'] == -1), 'class_pm10'] = 3

df_daily.head()
```

```python
# 필요한 모듈 불러오기
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
import pandas as pd
```

```python
# 널 값 채우기 
df_daily['Daejeon_Oil'] = df_daily['Daejeon_Oil'].fillna(0)
df_daily['Ulsan_Oil'] = df_daily['Ulsan_Oil'].fillna(0)
df_daily['Jeju_LNG'] = df_daily['Jeju_LNG'].fillna(0)
```

# target 라벨링, 슬라이딩 윈도우 적용


```python
#target 값인 Seoul_PM2.5에 대해 0,1,2,3 라벨링 해주는 코드 셀

df = df_daily.dropna(subset=['Seoul_PM2.5','Seoul_PM10']).reset_index(drop = True) 
#target인 Seoul_PM2.5/10에 nan 값 있는 행 삭제, index 초기화

pm25 = [16, 36, 76]
pm10 = [31, 81, 151]

df['class_pm2.5'] = -1
df['class_pm10'] = -1

for i,pm in enumerate(pm25):
  print(i,pm)
  df.loc[(df['class_pm2.5'] == -1) & (df['Seoul_PM2.5'] < pm), 'class_pm2.5'] = i
df.loc[(df['class_pm2.5'] == -1), 'class_pm2.5'] = 3

for i,pm in enumerate(pm10):
  df.loc[(df['class_pm10'] == -1) & (df['Seoul_PM10'] < pm), 'class_pm10'] = i
df.loc[(df['class_pm10'] == -1), 'class_pm10'] = 3

```

```python
#특히 중국은 당일의 미세먼지보다 하루 이틀 전의 미세먼지가 한국에 영향을 미칠 것으로 예상되니까
#타겟 데이터 기준 1일 전, 2일 전 데이터 생성 -> 이전 날들의 미세먼지 기록을 반영할 수 있도록 한다.


Seoul_PM25 = pd.DataFrame({'Seoul_PM2.5(t-2)':[],'Seoul_PM2.5(t-1)':[]})
Seoul_PM10 = pd.DataFrame({'Seoul_PM10(t-2)':[],'Seoul_PM10(t-1)':[]})
Beijing_PM10 = pd.DataFrame({'Beijing_PM10(t-2)':[],'Beijing_PM10(t-1)':[]})
Beijing_PM25 = pd.DataFrame({'Beijing_PM2.5(t-2)':[],'Beijing_PM2.5(t-1)':[]})
Seoul_PMtarget = pd.DataFrame({'Seoul_PM2.5(t)':[],'Seoul_PM10(t)':[]})

temp = df_daily.copy()

temp.reset_index(drop=True, inplace=True)

for i in range(temp.shape[0]-2):
  rowS25 = pd.DataFrame({'Seoul_PM2.5(t-2)':[temp['Seoul_PM2.5'][i]],'Seoul_PM2.5(t-1)':[temp['Seoul_PM2.5'][i+1]]})
  rowS10 = pd.DataFrame({'Seoul_PM10(t-2)':[temp['Seoul_PM10'][i]],'Seoul_PM10(t-1)':[temp['Seoul_PM10'][i+1]]})
  rowB25 = pd.DataFrame({'Beijing_PM2.5(t-2)':[temp['Beijing_PM2.5'][i]],'Beijing_PM2.5(t-1)':[temp['Beijing_PM2.5'][i+1]]})
  rowB10 = pd.DataFrame({'Beijing_PM10(t-2)':[temp['Beijing_PM10'][i]],'Beijing_PM10(t-1)':[temp['Beijing_PM10'][i+1]]})
  target = pd.DataFrame({'Seoul_PM2.5(t)':[temp['Seoul_PM2.5'][i+2]],'Seoul_PM10(t)':[temp['Seoul_PM10'][i+2]]})

  Seoul_PM25 = pd.concat([Seoul_PM25, rowS25], ignore_index = True)
  Seoul_PM10 = pd.concat([Seoul_PM10, rowS10], ignore_index = True)
  Beijing_PM25 = pd.concat([Beijing_PM25, rowB25], ignore_index = True)
  Beijing_PM10 = pd.concat([Beijing_PM10, rowB10], ignore_index = True)
  Seoul_PMtarget = pd.concat([Seoul_PMtarget, target], ignore_index = True)



```

## train_test_split_pm2.5


```python
# pm 2.5 input data, target data 생성 
input = df.loc[:, ['Beijing_PM2.5(t-2)', 'Beijing_PM2.5(t-1)',
       'Seoul_NO2', 'Seoul_O3','Seoul_CO', 'Seoul_SO2', 'GW_LNG', 'GW_Coal', 
       'GG_LNG', 'GG_Coal', 'GG_Oil', 'GN_Coal', 'GN_Oil', 'GB_LNG', 'GB_Coal', 'Gwangju_LNG',
       'Daegu_LNG', 'Daegu_Coal', 'Daegu_Oil', 'Daejeon_LNG', 'Daejeon_Oil',
       'Busan_LNG', 'Busan_Coal', 'Seoul_LNG', 'Sejong_LNG', 'Ulsan_LNG',
       'Ulsan_Oil', 'Incheon_LNG', 'Incheon_Coal', 'JN_LNG', 'JN_Coal',
       'JB_LNG', 'JB_Coal', 'JB_Oil', 'Jeju_Oil', 'CN_LNG', 'CN_Coal',
       'CN_Oil', 'CB_Oil', 'Jeju_LNG']].to_numpy()
target = df.iloc[:, -4].to_numpy()
```

```python
# 데이터 분할하기
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(input, target, test_size = 0.2, random_state=777) 
```

```python
# 정규화
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)

train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import cross_validate
import numpy as np

# XGBoost 회귀 모델 생성
xgb = XGBRegressor(random_state=42)
scores = cross_validate(xgb, train_scaled, train_target, return_train_score=True, n_jobs=-1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))

# 과대 적합 -> 그리드 서치, 랜덤 서치
```

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor
from sklearn.model_selection import cross_validate
import numpy as np

xgb1=XGBRegressor(random_state=42)

xgb_param_grid={'max_depth' : [3, 4, 5, 6, 7, 8],
                'n_estimators' : [100, 200, 400],
                'learning_rate' : [0.1, 0.2]}

xgb_grid=GridSearchCV(xgb1, param_grid=xgb_param_grid, verbose=2, n_jobs=-1)
xgb_grid.fit(train_scaled, train_target)
print(xgb_grid.best_params_)
print(np.max(xgb_grid.cv_results_['mean_test_score']))
xgbgs = xgb_grid.best_estimator_
print(xgbgs.score(train_scaled, train_target))
print(xgbgs.score(test_scaled, test_target))
```

```python
from scipy.stats import uniform, randint
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBRegressor
import numpy as np

xgb2=XGBRegressor(random_state=42)

xgb_param_random={'max_depth' : randint(3, 10),
                  'n_estimators' : randint(100, 500),
                  'learning_rate' : uniform(0.01, 0.15)}

xgb_random=RandomizedSearchCV(xgb2, xgb_param_random, verbose=2, n_jobs=-1)
xgb_random.fit(train_scaled, train_target)
print(xgb_random.best_params_)
print(np.max(xgb_random.cv_results_['mean_test_score']))
xgbrs = xgb_random.best_estimator_

# 훈련 데이터 셋, 테스트 데이터 셋 성능 확인
print(xgbrs.score(train_scaled, train_target))
print(xgbrs.score(test_scaled, test_target))
```
