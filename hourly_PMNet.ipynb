# target 라벨링, 슬라이딩 윈도우 적용


```python
#target 값인 Seoul_PM2.5에 대해 0,1,2,3 라벨링 해주는 코드 셀

df = df.dropna(subset=['Seoul_PM2.5','Seoul_PM10']).reset_index(drop = True) 
#target인 Seoul_PM2.5/10에 nan 값 있는 행 삭제, index 초기화

pm25 = [16, 36, 76]
pm10 = [31, 81, 151]

df['class_pm2.5'] = -1
df['class_pm10'] = -1

for i,pm in enumerate(pm25):
  print(i,pm)
  df.loc[(df['class_pm2.5'] == -1) & (df['Seoul_PM2.5'] < pm), 'class_pm2.5'] = i
df.loc[(df['class_pm2.5'] == -1), 'class_pm2.5'] = 3

for i,pm in enumerate(pm10):
  df.loc[(df['class_pm10'] == -1) & (df['Seoul_PM10'] < pm), 'class_pm10'] = i
df.loc[(df['class_pm10'] == -1), 'class_pm10'] = 3

```

# 모델링


```python
import pandas as pd
import numpy as np
import tensorflow as tf
from keras.datasets import mnist
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler, RobustScaler
```

```python
#train set과 test set을 각각 목적에 맞게 pm2.5/pm10 분류/회귀 로 split --> 25c,25r,10c,10r
#random_state에 같은 값을 줘 X_train과 X_test는 전부 같다

X_trainl, X_test, y_train_25c, y_test_25c = train_test_split(X, y_25_cla, test_size = 0.2, random_state = 42)
X_trainl, X_test, y_train_25r, y_test_25r = train_test_split(X, y_25_reg, test_size = 0.2, random_state = 42)
X_trainl, X_test, y_train_10c, y_test_10c = train_test_split(X, y_10_cla, test_size = 0.2, random_state = 42)
X_trainl, X_test, y_train_10r, y_test_10r = train_test_split(X, y_10_reg, test_size = 0.2, random_state = 42)

X_train, X_val, y_train_25c, y_val_25c = train_test_split(X_trainl, y_train_25c, test_size = 0.2, random_state = 42)
X_train, X_val, y_train_25r, y_val_25r = train_test_split(X_trainl, y_train_25r, test_size = 0.2, random_state = 42)
X_train, X_val, y_train_10c, y_val_10c = train_test_split(X_trainl, y_train_10c, test_size = 0.2, random_state = 42)
X_train, X_val, y_train_10r, y_val_10r = train_test_split(X_trainl, y_train_10r, test_size = 0.2, random_state = 42)
```

```python
print(X_train.shape, X_val.shape, X_test.shape)
```

```python
#여러 scaler 적용, 딱히 별 차이 없었음.
scaler = StandardScaler()
#scaler = MinMaxScaler()
#scaler = MaxAbsScaler()
#scaler = RobustScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_val = scaler.transform(X_val)
```

```python
# 모델에서 사용할 기본적인 block 정의.
# block은 같은 크기의 layer 2개를 포함하며, 각 layer 사이에는 Dropout과 Batch Normalization이 적용된다.
# 또한, activation은 relu, elu, leakyrelu 중 원하는 것을 선택할 수 있도록 했다.

class Block(tf.keras.layers.Layer):
  def __init__(self, units, activation, initializer='he_uniform'):
    super(Block, self).__init__()
    # 필요한 layer 선언 
    self.fc1 = tf.keras.layers.Dense(units, activation = None, kernel_initializer = initializer)
    self.drop = tf.keras.layers.Dropout(0.25)
    self.fc2 = tf.keras.layers.Dense(units, activation = None, kernel_initializer = initializer)
    self.bn1 = tf.keras.layers.BatchNormalization()
    self.bn2 = tf.keras.layers.BatchNormalization()

    #입력 받은 activation에 따라 해당 block에서 사용할 activation layer 선언
    if activation == 'relu':
      self.activation = tf.keras.layers.ReLU()
    elif activation == 'elu':
      self.activation = tf.keras.layers.ELU()
    elif activation == 'leakyrelu':
      self.activation = tf.keras.layers.LeakyReLU() 

  def __call__(self, inputs):
    #block의 흐름에 맞게 layer를 호출
    out = self.fc1(inputs)
    out = self.bn1(out)
    out = self.drop(out)
    out = self.activation(out)

    out = self.fc2(out)
    out = self.bn2(out)
    out = self.drop(out)
    out = self.activation(out)
    return out


# 위에서 정의한 클래스 Block을 이용해 모델 PMNet 정의
class PMNet(tf.keras.Model):
    def __init__(self, num_classes) -> None:
        super(PMNet, self).__init__()
        self.inp = tf.keras.layers.InputLayer(input_shape=(43,),batch_size=32)
        #필요한 block들을 선언한다. 
        # 총 64-64-128-128-256-256-512-512로 layer를 쌓을 예정.
        self.block1 = Block(64, 'relu')
        self.block2 = Block(128, 'leakyrelu')
        self.block3 = Block(256, 'leakyrelu')
        self.block4 = Block(512, 'leakyrelu')
        self.drop = tf.keras.layers.Dropout(0.5)

        self.outp = tf.keras.layers.Dense(num_classes, activation="softmax")


    def call(self, x):
      # 순서대로 block 쌓아서 모델을 구성한다.
        x = self.inp(x)
        x = self.block1(x)
        x = self.block2(x)
        x = self.block3(x)
        x = self.block4(x)
        x = self.drop(x)
        x = self.outp(x)
        return x
  
```

# Pm 10 classification

```python
model10 = PMNet(num_classes=4)
#모델 선언 후 optimizer, loss function, metric 지정하고 학습 진행

model10.compile(optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])


history_10c = model10.fit(X_train, y_train_10c,
          batch_size=32,
          epochs=150,
          verbose=1,
          validation_data=(X_val, y_val_10c))
```

```python
import matplotlib.pyplot as plt
#model.fit하면서 선언했던 history 변수를 이용한다.
#history.history는 딕셔너리 형식으로 되어 있고, loss, accuracy, val_loss, val_accuracy를 key로 그에 해당하는 학습 진행 중 값들을 저장한다.

#matplotlib을 통해 epoch 진행에 따른 train_accuracy와 val_accuracy 그리기
plt.plot(history_10c.history['accuracy'])
plt.plot(history_10c.history['val_accuracy'])

#x축과 y축에 붙일 이름 지정.
plt.xlabel('epoch')
plt.ylabel('accuracy')
#범례 표시
plt.legend(['train_accuracy', 'val_accuracy'])
#show하면서 비워지고, 다음으로 loss 그리기
plt.show()

#epoch 진행에 따른 train_loss와 val_loss 그리기
plt.plot(history_10c.history['loss'])
plt.plot(history_10c.history['val_loss'])
#x축과 y축에 붙일 이름 지정.
plt.xlabel('epoch')
plt.ylabel('loss')
#범례 표시
plt.legend(['train', 'val'])
plt.show()



```

```python
model10 = PMNet(num_classes=4)
#모델 선언 후 optimizer, loss function, metric 지정하고 학습 진행

model10.compile(optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])


history_10c = model10.fit(X_train, y_train_10c,
          batch_size=32,
          epochs=65,
          verbose=1,
          validation_data=(X_val, y_val_10c))
```

```python
y_pred = model10.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test_10c, np.argmax(y_pred, axis=1)))
```
