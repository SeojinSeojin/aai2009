## 일단 일별 평균 데이터만 사용하기

```python
df_daily = df.groupby(['Year','Month','Day']).mean() # 일별 평균값. NaN 값들은 평균 계산에 포함되지 않는다.
df_daily = df_daily.drop('UTCHour', axis=1)
```

# target 라벨링, 슬라이딩 윈도우 적용


```python
#target 값인 Seoul_PM2.5에 대해 0,1,2,3 라벨링 해주는 코드 셀

df = df_daily.dropna(subset=['Seoul_PM2.5','Seoul_PM10']).reset_index(drop = True) 
#target인 Seoul_PM2.5/10에 nan 값 있는 행 삭제, index 초기화

pm25 = [16, 36, 76]
pm10 = [31, 81, 151]

df['class_pm2.5'] = -1
df['class_pm10'] = -1

for i,pm in enumerate(pm25):
  print(i,pm)
  df.loc[(df['class_pm2.5'] == -1) & (df['Seoul_PM2.5'] < pm), 'class_pm2.5'] = i
df.loc[(df['class_pm2.5'] == -1), 'class_pm2.5'] = 3

for i,pm in enumerate(pm10):
  df.loc[(df['class_pm10'] == -1) & (df['Seoul_PM10'] < pm), 'class_pm10'] = i
df.loc[(df['class_pm10'] == -1), 'class_pm10'] = 3

```

```python
#타겟 데이터 기준 1일 전, 2일 전 데이터 생성 -> 이전 날들의 미세먼지 기록을 반영할 수 있도록. 
#특히 중국은 당일의 미세먼지보다 하루 이틀 전의 미세먼지가 한국에 영향을 미칠 것으로 예상되니까

Seoul_PM25 = pd.DataFrame({'Seoul_PM2.5(t-2)':[],'Seoul_PM2.5(t-1)':[]})
Seoul_PM10 = pd.DataFrame({'Seoul_PM10(t-2)':[],'Seoul_PM10(t-1)':[]})
Beijing_PM10 = pd.DataFrame({'Beijing_PM10(t-2)':[],'Beijing_PM10(t-1)':[]})
Beijing_PM25 = pd.DataFrame({'Beijing_PM2.5(t-2)':[],'Beijing_PM2.5(t-1)':[]})
Seoul_PMtarget = pd.DataFrame({'Seoul_PM2.5(t)':[],'Seoul_PM10(t)':[]})

temp = df_daily.copy()

temp.reset_index(drop=True, inplace=True)

for i in range(temp.shape[0]-2):
  rowS25 = pd.DataFrame({'Seoul_PM2.5(t-2)':[temp['Seoul_PM2.5'][i]],'Seoul_PM2.5(t-1)':[temp['Seoul_PM2.5'][i+1]]})
  rowS10 = pd.DataFrame({'Seoul_PM10(t-2)':[temp['Seoul_PM10'][i]],'Seoul_PM10(t-1)':[temp['Seoul_PM10'][i+1]]})
  rowB25 = pd.DataFrame({'Beijing_PM2.5(t-2)':[temp['Beijing_PM2.5'][i]],'Beijing_PM2.5(t-1)':[temp['Beijing_PM2.5'][i+1]]})
  rowB10 = pd.DataFrame({'Beijing_PM10(t-2)':[temp['Beijing_PM10'][i]],'Beijing_PM10(t-1)':[temp['Beijing_PM10'][i+1]]})
  target = pd.DataFrame({'Seoul_PM2.5(t)':[temp['Seoul_PM2.5'][i+2]],'Seoul_PM10(t)':[temp['Seoul_PM10'][i+2]]})

  Seoul_PM25 = pd.concat([Seoul_PM25, rowS25], ignore_index = True)
  Seoul_PM10 = pd.concat([Seoul_PM10, rowS10], ignore_index = True)
  Beijing_PM25 = pd.concat([Beijing_PM25, rowB25], ignore_index = True)
  Beijing_PM10 = pd.concat([Beijing_PM10, rowB10], ignore_index = True)
  Seoul_PMtarget = pd.concat([Seoul_PMtarget, target], ignore_index = True)



```

# 모델링


```python
import pandas as pd
import numpy as np
import tensorflow as tf
from keras.datasets import mnist
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from tensorflow import keras
```

```python
#train set과 test set을 각각 목적에 맞게 pm2.5/pm10 분류/회귀 로 split --> 25c,25r,10c,10r
#random_state에 같은 값을 줘 X_train과 X_test는 전부 같다

X_trainl, X_test, y_train_25c, y_test_25c = train_test_split(X, y_25_cla, test_size = 0.2, random_state = 42)
X_trainl, X_test, y_train_25r, y_test_25r = train_test_split(X, y_25_reg, test_size = 0.2, random_state = 42)
X_trainl, X_test, y_train_10c, y_test_10c = train_test_split(X, y_10_cla, test_size = 0.2, random_state = 42)
X_trainl, X_test, y_train_10r, y_test_10r = train_test_split(X, y_10_reg, test_size = 0.2, random_state = 42)

X_train, X_val, y_train_25c, y_val_25c = train_test_split(X_trainl, y_train_25c, test_size = 0.2, random_state = 42)
X_train, X_val, y_train_25r, y_val_25r = train_test_split(X_trainl, y_train_25r, test_size = 0.2, random_state = 42)
X_train, X_val, y_train_10c, y_val_10c = train_test_split(X_trainl, y_train_10c, test_size = 0.2, random_state = 42)
X_train, X_val, y_train_10r, y_val_10r = train_test_split(X_trainl, y_train_10r, test_size = 0.2, random_state = 42)
```

# Pm 10 classification

```python
from tensorflow import keras
from sklearn.preprocessing import StandardScaler
import numpy as np

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)

X_train_reshaped = X_train_scaled.reshape(X_train_scaled.shape[0], 1, X_train_scaled.shape[1])
X_val_reshaped = X_val_scaled.reshape(X_val_scaled.shape[0], 1, X_val_scaled.shape[1])

y_train_categorical = keras.utils.to_categorical(y_train_25c, num_classes=4)
y_val_categorical = keras.utils.to_categorical(y_val_25c, num_classes=4)

model = keras.Sequential()
model.add(keras.layers.LSTM(units=64, activation='relu', return_sequences=True, input_shape=(1, X_train_scaled.shape[1])))
model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.LSTM(units=32, activation='relu', return_sequences=True))
model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.LSTM(units=16, activation='relu'))
model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.Dense(units=16, activation='relu', kernel_regularizer=keras.regularizers.l2(0.001)))
model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.Dense(units=4, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Add early stopping callback
early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model.fit(X_train_reshaped, y_train_categorical, epochs=150, batch_size=32, validation_data=(X_val_reshaped, y_val_categorical), callbacks=[early_stopping])

score = model.evaluate(X_val_reshaped, y_val_categorical)
print("Validation Loss:", score[0])
print("Validation Accuracy:", score[1])
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
X_test_scaled = scaler.transform(X_test)
X_test_reshaped = X_test_scaled.reshape(X_test_scaled.shape[0], 1, X_test_scaled.shape[1])

y_pred = model.predict(X_test_reshaped)
```
